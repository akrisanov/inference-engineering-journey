# Linear Combinations of 2D Vectors

```python
import torch
```

```python
# Define two 2D vectors
v1 = torch.tensor([2.0, -1.0])
v2 = torch.tensor([1.0, 3.0])
```

```python
# Variable coefficients (scalars)
a = 4.0
b = -2.0
```

```python
# Linear combination: a*v1 + b*v2
combination = a * v1 + b * v2
```

```python
print(f"a * v1 + b * v2 = {combination}")
```

### How it works

- $a$ and $b$ are **scalar coefficients**
- $v_1$ and $v_2$ are 2D vectors

The formula for a linear combination is: $\vec u = a*\vec{v_1} + b*\vec{v_2}$

### Plotting with `matplotlib`

```python
import matplotlib.pyplot as plt
```

```python
# Plot setup
fig, ax = plt.subplots(figsize=(6, 6))
ax.set_aspect('equal')
ax.grid(True, which='both', linestyle='--', alpha=0.6)
ax.axhline(0, color='black', lw=1)
ax.axvline(0, color='black', lw=1)

# Origin point
origin = torch.tensor([0.0, 0.0])

# Plot original vectors
ax.quiver(*origin, *v1, angles='xy', scale_units='xy', scale=1, color='blue', label='v1')
ax.quiver(*origin, *v2, angles='xy', scale_units='xy', scale=1, color='green', label='v2')

# Plot scaled vectors
ax.quiver(*origin, *(a * v1), angles='xy', scale_units='xy', scale=1, color='blue', alpha=0.5, label=f'{a} * v1')
ax.quiver(*origin, *(b * v2), angles='xy', scale_units='xy', scale=1, color='green', alpha=0.5, label=f'{b} * v2')

# Plot linear combination result
ax.quiver(*origin, *combination, angles='xy', scale_units='xy', scale=1, color='red', label='a*v1 + b*v2')

# Axis limits and labels
ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10)
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.legend()
ax.set_title(f"Linear Combination: {a}*v1 + {b}*v2 = {tuple(combination.numpy())}")
```
